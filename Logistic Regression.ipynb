```python
#Logistic Regression is type of classification
# Finding the Median

data_preprocessed ['Absenteeism Time in Hours'].median()


#Since the median is 3, then everyone absent above the median of 3 hours is considered excessively absent
```

```python
#To find the targets or to classify the data as excessive or not - you have to ID them as 0's or 1's.
# can also use 
# targets = np.where (data_preprocessed ['Absenteeism Time in Hours']>3, 1, 0)

targets = np.where (data_preprocessed ['Absenteeism Time in Hours']> data_preprocessed ['Absenteeism Time in Hours'].median(), 1, 0)
```

```python
data_preprocessed ['Excessive Absenteeism'] = targets
```
